Using the information in references I learned about the relation ship of friction and momentum. It produces an exponential relationship as the ball loses momentum over time. To simulate this I decreased the ball momentum during each iteration of the loop by a small percentage (that varies based on rate and time step size), in order to best recreate the force of friction on the ball. Since this value is also related to gravity, at a certain threshold movement no longer becomes possible with the remaining momentum, causing the ball the stay in place. I simulated this by checking the momentum of each ball during each iteration of the loop. Once all balls had crossed this threshold (where they would no long realistically be moving) their momentums are set to zero, and the cue ball is hit again.
